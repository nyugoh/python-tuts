```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1, 2, 4,6], [0, 1, 2, 5], 'b--', label='sodas')
plt.plot([1, 4, 4,9], [0, 3, 2, 7], 'c--', label='milk')
plt.legend(loc='best')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(np.arange(4), np.arange(4))
ax2 = fig.add_subplot(2, 2, 2)

ax2.plot(np.random.randn(50).cumsum(), 'c--')
ax3 = fig.add_subplot(2, 2, 3)
```

```python
plt.plot(np.random.randn(30).cumsum(), 'bo--')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.plot(np.random.randn(50).cumsum(), 'ko--')
ax1.set_title('Some rand data')
ax1.set_xlabel('rands')
ax1.set_ylabel('time')
ax1.set_xticks([5, 15, 27, 35, 45])
ax1.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30)
fig.show()
```
