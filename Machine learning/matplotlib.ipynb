```python
import matplotlib.pyplot as plt

x = [2, 4, 6, 8]
y = [10, 20, 30, 50]

x2 = [1, 4, 5, 9]
y2 = [7, 35, 40, 55]

plt.plot(x, y, label='Milk')
plt.plot(x2, y2, label='Soda')

plt.title("Test example")
plt.xlabel('Time')
plt.ylabel('Litres')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x1 = [1, 3, 5, 7, 9]
y1 = [10, 20, 30, 50, 60]

x2 = [0, 2, 4, 6, 8]
y2 = [5, 15, 25, 35, 45]

plt.bar(x1, y1, label='Sodas', color='b')
plt.bar(x2, y2, label='Milk', color='orange')

plt.xlabel('Week days')
plt.ylabel('Litres')
plt.legend()
plt.show()
```
