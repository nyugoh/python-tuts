# Import tools

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

## Import data

```python
data = pd.read_csv('./data/Pokemon.csv', index_col=0)
print(data.head())
```

```python
sns.lmplot(x='Attack', y='Defense', data=data)
```

```python
# Swarm plot with Pokemon color palette
sns.swarmplot(x='Type 1', y='Attack', data=data)
```

```python
sns.distplot(data.Attack)
```

```python
plt.bar(data.Attack, data.Defense)
plt.show()
```

```python
sns.kdeplot(data.Attack, data.Defense)
```
